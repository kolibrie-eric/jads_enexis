```python
# overlay en area
# https://gis.stackexchange.com/questions/375844/removing-one-to-many-links-when-performing-spatial-join-in-geopandas
```

```python
def get_largest_area(df1, df2, groupby):
    overlay = gpd.overlay(df1, df2)
    #Calculate the area and sort descending
    overlay['area'] = overlay.area
    overlay.sort_values('area', ascending=False, inplace=True)
    overlay = overlay.groupby(groupby, as_index=False).first()
    return overlay

pc4_res = get_largest_area(pc4, res, 'PC4')
```

```python
# alleen de noodzakelijke columns behouden
pc4_res = pc4_res[['PC4','Regio']]
pc4_res = pc4_res.rename(columns={'Regio': 'RES'})
```

```python
# check missing values
def missing_values(df):
    missing_total = df.isna().sum().sort_values(ascending=False)
    print('Number of attributes with Null values: ' + str(len(missing_total[missing_total > 0])))
    print(missing_total[missing_total > 0])

missing_values(pc4_res)
```

```python
def count_values(df):
    col = df.columns
    for c in col:
        print('')
        print(f'Column: {c}')
        print(df[c].value_counts())

count_values(pc4_res)
```
